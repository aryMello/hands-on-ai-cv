**TASK: Use PIL and matplotlib to read and display the ../DATA/00-puppy.jpg image.**

```python
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
img = Image.open("../DATA/00-puppy.jpg")

# Display the image using matplotlib
plt.imshow(img)
plt.axis('off')  # hides the axis for better image display
plt.show()
```

**TASK: Convert the image to a NumPy Array**

```python
img_array = np.array(img)
img_array.shape
```

**FINAL TASK: Use slicing to set the RED and GREEN channels of the picture to 0, then use imshow() to show the isolated blue channel**

```python
pic_blue = img_array.copy()
```

```python
pic_blue[:, :, 0] = 0  # Red channel
pic_blue[:, :, 1] = 0  # Green channel
```

```python
plt.imshow(pic_blue)
```
